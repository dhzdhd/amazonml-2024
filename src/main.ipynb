```python
import pytesseract
from PIL import Image
from transformers import pipeline
from transformers import AutoTokenizer, AutoModelForTokenClassification
import os
```

```python
TESSERACT_PATH = r"C:\Program Files\Tesseract-OCR\tesseract"
IMAGES_PATH = "../images"
```

```python
pytesseract.pytesseract.tesseract_cmd = TESSERACT_PATH
```

```python
ocr = []

for image in os.listdir(IMAGES_PATH):
    raw = pytesseract.image_to_string(Image.open(f"{IMAGES_PATH}/{image}"))
    raw = raw.lower().replace("\n", " ")
    ocr.append(raw)
```

```python
tokenizer = AutoTokenizer.from_pretrained("Babelscape/cner-base")
model = AutoModelForTokenClassification.from_pretrained("Babelscape/cner-base")
```

```python
ner = pipeline("ner", model=model, tokenizer=tokenizer, grouped_entities=True, device="cuda")
```

```python
def contains_numeric(word: str) -> bool:
    for char in list(word):
        if char.isdigit():
            return True
    return False
```

```python
ner_result = []

for image in ocr:
    result = ner(image)
    result = list(filter(lambda x: x['entity_group'] == 'MEASURE', result))
    result = list(filter(lambda x: contains_numeric(x['word']), result))

    ner_result.append(result)
```
